```python
def get_hash_table_data(key):
    if key in AiportTimeWeatherHash.keys():
        return AiportTimeWeatherHash[key]
    else:
        return {}
```

```python
def get_min_index(time_array,time):
    FMT = '%H:%M:%S'
    dict = {}
    min_index = 0
    for items in time_array:
        tdelta = abs(datetime.strptime((str(items['time'])+':00'), FMT) - datetime.strptime(time, FMT)).seconds
        dict[items['index']] = tdelta
    return (min(dict.items(), key=lambda x: x[1])[0])
    
                
                
```

```python
flight_data = pd.read_csv('/home/kkanagalananthapadm1/Flight_data_for_modelling.csv')
flight_data
```

```python
Temp = []
Conditions = []
Visibility = []
Wind_Speed = []
count = 0
for index,rows in flight_data.iterrows():
    date = pd.to_datetime(rows['SCHEDULED_DEPARTURE']).date()
    time = str(pd.to_datetime(rows['SCHEDULED_DEPARTURE']).time())
    key = rows['ORIGIN_AIRPORT'] + str(date)
    time_array = get_hash_table_data(key)
    if time_array:
        mindate_index = get_min_index(time_array,time)
        Temp.append(weather_copy_ref.ix[mindate_index]['Temp'])
        Conditions.append(weather_copy_ref.ix[mindate_index]['Conditions'])
        Visibility.append(weather_copy_ref.ix[mindate_index]['Visibility'])
        Wind_Speed.append(weather_copy_ref.ix[mindate_index]['Wind_speed'])
    else:
        Temp.append(9999)
        Conditions.append('Not Known')
        Visibility.append(99)
        Wind_Speed.append(0)
    count+=1
    if count%10000 == 0:
        print (count)
        
flight_data['Temp'] = Temp
flight_data['Conditions'] = Conditions
flight_data['Visibility'] = Visibility
flight_data['Wind_speed'] = Wind_Speed
```

```python
def getTimeForEachDay(weatherOnDayTime):
    date_iso = date(int(weatherOnDayTime['Year']), int(weatherOnDayTime['Month']), int(weatherOnDayTime['Day'])).isoformat()
    airport_date_iso = weatherOnDayTime['Airport'] + str(date_iso)
    #print(airport_date_iso)
    #print(weatherOnDayTime['index'])
    if airport_date_iso in list(AiportTimeWeatherHash.keys()):
        #print("inside if")
        #print(AiportTimeWeatherHash[airport_date_iso])
        AiportTimeWeatherHash[airport_date_iso] += [{'index': weatherOnDayTime['index'], 'time': weatherOnDayTime['Time']}]
        #print(AiportTimeWeatherHash[airport_date_iso])
        #print("\n\n")
        #return AiportTimeWeatherHash[airport_date_iso]
    else:
        #print("inside else")
        AiportTimeWeatherHash[airport_date_iso] = [{'index': weatherOnDayTime['index'], 'time': weatherOnDayTime['Time']}]
        #print(AiportTimeWeatherHash[airport_date_iso])
        #print("\n\n")
        #return AiportTimeWeatherHash[airport_date_iso]
    #print(AiportTimeWeatherHash[airport_date_iso])
```

```python
AiportTimeWeatherHash = {}
weather_copy_ref[['Airport','Year','Month','Day','Time','index']].apply(lambda x: getTimeForEachDay(x), axis=1)
```

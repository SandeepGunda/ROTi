```python
# Selecting the needed variables from the dataframe for training and testing set.
tr_data = df.iloc[:,[0,1,2,3,5,6,7,8,9,11,12,13,14,15,16]]
tr_target = df.iloc[:,10]

print tr_data.head
print tr_target.head
```
